Netflix Data Analysis

This project analyzes the Netflix Movies and TV Shows dataset to discover trends in content production, genres, and countries.

Tools Used:
- Python
- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
df = df.dropna(subset=['country','rating'])
```

```python
sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows on Netflix")

plt.show()
```

Insight: Netflix contains significantly more movies than TV shows.

```python
df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top Countries Producing Netflix Content")

plt.show()
```

Insight: The United States produces the majority of Netflix titles.

```python
df['release_year'].value_counts().sort_index().plot()

plt.title("Netflix Content Growth Over Years")

plt.show()
```

Insight: Netflix content production increased significantly after 2015.

```python
genres = df['listed_in'].str.split(',', expand=True).stack()

genres.value_counts().head(10).plot(kind='bar')

plt.title("Top Genres on Netflix")

plt.show()
```

Insight: Drama and Comedy dominate Netflix's catalog.

## Conclusion

From the analysis:

- Netflix has significantly more movies than TV shows
- The USA produces the most Netflix content
- Drama and Comedy are the most popular genres
- Content production increased rapidly after 2015
